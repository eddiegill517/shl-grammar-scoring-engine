```python
import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import gc

import librosa
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import KFold, RepeatedKFold
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr
from scipy.optimize import minimize

import whisper
from transformers import (
    AutoTokenizer, AutoModel, AutoModelForSequenceClassification,
    RobertaTokenizer, RobertaForSequenceClassification
)

SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {DEVICE}")

BASE_PATH = '/kaggle/input/shl-intern-hiring-assessment-2025/dataset'

TRAIN_AUDIO_PATH = os.path.join(BASE_PATH, 'audios/train')
TEST_AUDIO_PATH = os.path.join(BASE_PATH, 'audios/test')
CSV_PATH = os.path.join(BASE_PATH, 'csvs')

train_df = pd.read_csv(os.path.join(CSV_PATH, 'train.csv'))
test_df = pd.read_csv(os.path.join(CSV_PATH, 'test.csv'))
print(f"Train: {len(train_df)}, Test: {len(test_df)}")


class GrammarScorer:
    """Uses CoLA-trained model for grammatical acceptability scoring"""
    
    def __init__(self):
        print("Loading Whisper...")
        self.whisper = whisper.load_model("base")
        
        print("Loading CoLA grammar model...")
        self.cola_tokenizer = AutoTokenizer.from_pretrained("textattack/roberta-base-CoLA")
        self.cola_model = AutoModelForSequenceClassification.from_pretrained("textattack/roberta-base-CoLA")
        self.cola_model.to(DEVICE)
        self.cola_model.eval()
        
        print("Loading language model for embeddings...")
        self.lm_tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-small")
        self.lm_model = AutoModel.from_pretrained("microsoft/deberta-v3-small")
        self.lm_model.to(DEVICE)
        self.lm_model.eval()
        
        print("All models loaded!")
    
    def transcribe(self, audio_path):
        try:
            result = self.whisper.transcribe(audio_path, language='en')
            return result['text'].strip()
        except:
            return ""
    
    def get_cola_score(self, text):
        """Get grammatical acceptability score from CoLA model"""
        if not text or len(text) < 5:
            return 0.5, 0.5
        
        try:
            sentences = [s.strip() for s in text.replace('!', '.').replace('?', '.').split('.') if len(s.strip()) > 3]
            if not sentences:
                return 0.5, 0.5
            
            scores = []
            for sent in sentences[:30]:
                inputs = self.cola_tokenizer(sent, return_tensors="pt", truncation=True, max_length=128, padding=True)
                inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
                
                with torch.no_grad():
                    outputs = self.cola_model(**inputs)
                    probs = torch.softmax(outputs.logits, dim=-1)
                    acceptable_prob = probs[0, 1].item()
                    scores.append(acceptable_prob)
            
            return np.mean(scores), np.std(scores) if len(scores) > 1 else 0
        except:
            return 0.5, 0.5
    
    def get_embeddings(self, text):
        """Get text embeddings from language model"""
        if not text:
            return np.zeros(384)
        
        try:
            inputs = self.lm_tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True)
            inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
            
            with torch.no_grad():
                outputs = self.lm_model(**inputs)
                hidden = outputs.last_hidden_state
                
                cls_emb = hidden[:, 0, :].cpu().numpy().flatten()
                mean_emb = hidden.mean(dim=1).cpu().numpy().flatten()
            
            combined = np.concatenate([cls_emb[:192], mean_emb[:192]])
            return combined
        except:
            return np.zeros(384)
    
    def extract_grammar_features(self, text):
        """Extract detailed grammar-related features"""
        if not text:
            return np.zeros(50)
        
        features = []
        words = text.split()
        sentences = [s.strip() for s in text.replace('!', '.').replace('?', '.').split('.') if s.strip()]
        
        features.append(len(words))
        features.append(len(sentences))
        features.append(len(words) / max(1, len(sentences)))
        
        word_lens = [len(w) for w in words]
        features.append(np.mean(word_lens) if word_lens else 0)
        features.append(np.std(word_lens) if len(word_lens) > 1 else 0)
        features.append(max(word_lens) if word_lens else 0)
        
        sent_lens = [len(s.split()) for s in sentences]
        features.append(np.mean(sent_lens) if sent_lens else 0)
        features.append(np.std(sent_lens) if len(sent_lens) > 1 else 0)
        features.append(max(sent_lens) if sent_lens else 0)
        features.append(min(sent_lens) if sent_lens else 0)
        
        features.append(len(set(words)) / max(1, len(words)))
        features.append(len(set(w.lower() for w in words)) / max(1, len(words)))
        
        long_words = sum(1 for w in words if len(w) > 6)
        features.append(long_words / max(1, len(words)))
        
        very_long = sum(1 for w in words if len(w) > 8)
        features.append(very_long / max(1, len(words)))
        
        features.append(text.count(','))
        features.append(text.count(',') / max(1, len(sentences)))
        features.append(text.count(';'))
        features.append(text.count(':'))
        features.append(text.count('?'))
        features.append(text.count('!'))
        
        filler_words = ['um', 'uh', 'like', 'you know', 'basically', 'actually', 'literally', 
                       'kind of', 'sort of', 'i mean', 'well']
        text_lower = text.lower()
        filler_count = sum(text_lower.count(f) for f in filler_words)
        features.append(filler_count)
        features.append(filler_count / max(1, len(words)))
        
        complex_conj = ['although', 'however', 'therefore', 'furthermore', 'moreover', 
                       'nevertheless', 'consequently', 'whereas', 'while', 'unless']
        features.append(sum(text_lower.count(c) for c in complex_conj))
        
        subordinate = ['because', 'since', 'although', 'though', 'while', 'when', 'if', 
                      'unless', 'after', 'before', 'until', 'as']
        features.append(sum(text_lower.count(s) for s in subordinate))
        
        relative = ['that', 'which', 'who', 'whom', 'whose', 'where', 'when']
        features.append(sum(text_lower.count(r) for r in relative))
        
        transitions = ['first', 'second', 'then', 'next', 'finally', 'also', 'additionally',
                      'in addition', 'for example', 'for instance', 'in fact', 'indeed']
        features.append(sum(text_lower.count(t) for t in transitions))
        
        contractions = ["n't", "'s", "'re", "'ve", "'ll", "'d", "'m"]
        features.append(sum(text.count(c) for c in contractions))
        features.append(sum(text.count(c) for c in contractions) / max(1, len(words)))
        
        proper_start = sum(1 for s in sentences if s and s[0].isupper())
        features.append(proper_start / max(1, len(sentences)))
        
        def count_syllables(word):
            word = word.lower()
            vowels = 'aeiou'
            count = 0
            prev_vowel = False
            for c in word:
                is_vowel = c in vowels
                if is_vowel and not prev_vowel:
                    count += 1
                prev_vowel = is_vowel
            return max(1, count)
        
        syllables = [count_syllables(w) for w in words]
        avg_syl = np.mean(syllables) if syllables else 1
        features.append(avg_syl)
        
        asl = len(words) / max(1, len(sentences))
        flesch_kincaid = 0.39 * asl + 11.8 * avg_syl - 15.59
        features.append(flesch_kincaid)
        
        flesch_reading = 206.835 - 1.015 * asl - 84.6 * avg_syl
        features.append(flesch_reading)
        
        complex_words = sum(1 for w in words if count_syllables(w) >= 3)
        features.append(complex_words / max(1, len(words)))
        
        articles = ['a', 'an', 'the']
        features.append(sum(1 for w in words if w.lower() in articles) / max(1, len(words)))
        
        prepositions = ['in', 'on', 'at', 'by', 'for', 'with', 'to', 'from', 'of', 'about']
        features.append(sum(1 for w in words if w.lower() in prepositions) / max(1, len(words)))
        
        pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us', 'them']
        features.append(sum(1 for w in words if w.lower() in pronouns) / max(1, len(words)))
        
        aux_verbs = ['is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had',
                    'do', 'does', 'did', 'will', 'would', 'could', 'should', 'may', 'might', 'must']
        features.append(sum(1 for w in words if w.lower() in aux_verbs) / max(1, len(words)))
        
        if len(words) > 1:
            bigrams = [f"{words[i]} {words[i+1]}" for i in range(len(words)-1)]
            features.append(len(set(bigrams)) / len(bigrams))
        else:
            features.append(1)
        
        if len(words) > 2:
            trigrams = [f"{words[i]} {words[i+1]} {words[i+2]}" for i in range(len(words)-2)]
            features.append(len(set(trigrams)) / max(1, len(trigrams)))
        else:
            features.append(1)
        
        features.append(len(text))
        features.append(sum(1 for c in text if c.isupper()) / max(1, len(text)))
        features.append(sum(1 for c in text if c.isdigit()) / max(1, len(text)))
        features.append(sum(1 for c in text if not c.isalnum() and c != ' ') / max(1, len(text)))
        
        return np.array(features[:50])
    
    def extract_audio_features(self, audio_path):
        """Extract fluency-related audio features"""
        try:
            y, sr = librosa.load(audio_path, sr=16000, mono=True, duration=60)
            features = []
            
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
            for i in range(13):
                features.extend([np.mean(mfcc[i]), np.std(mfcc[i])])
            
            rms = librosa.feature.rms(y=y)[0]
            features.extend([np.mean(rms), np.std(rms)])
            
            zcr = librosa.feature.zero_crossing_rate(y)[0]
            features.extend([np.mean(zcr), np.std(zcr)])
            
            tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
            features.append(float(tempo))
            
            onset_env = librosa.onset.onset_strength(y=y, sr=sr)
            onsets = librosa.onset.onset_detect(onset_envelope=onset_env, sr=sr)
            speech_rate = len(onsets) / (len(y) / sr)
            features.append(speech_rate)
            
            energy = librosa.feature.rms(y=y, frame_length=400, hop_length=160)[0]
            silence_threshold = np.mean(energy) * 0.1
            silence_ratio = np.sum(energy < silence_threshold) / len(energy)
            features.append(silence_ratio)
            
            is_silence = energy < silence_threshold
            changes = np.diff(is_silence.astype(int))
            num_pauses = np.sum(changes == 1)
            features.append(num_pauses)
            features.append(num_pauses / (len(y) / sr))
            
            pitches, magnitudes = librosa.piptrack(y=y, sr=sr, fmin=50, fmax=400)
            pitch_values = []
            for t in range(pitches.shape[1]):
                idx = magnitudes[:, t].argmax()
                if pitches[idx, t] > 0:
                    pitch_values.append(pitches[idx, t])
            
            if pitch_values:
                features.extend([np.mean(pitch_values), np.std(pitch_values), 
                               np.max(pitch_values) - np.min(pitch_values)])
            else:
                features.extend([0, 0, 0])
            
            return np.array(features[:40])
        except:
            return np.zeros(40)
    
    def extract_all(self, audio_path):
        text = self.transcribe(audio_path)
        
        cola_mean, cola_std = self.get_cola_score(text)
        embeddings = self.get_embeddings(text)
        grammar_feats = self.extract_grammar_features(text)
        audio_feats = self.extract_audio_features(audio_path)
        
        cola_features = np.array([cola_mean, cola_std, cola_mean * 5])
        
        all_features = np.concatenate([cola_features, embeddings, grammar_feats, audio_feats])
        all_features = np.nan_to_num(all_features, nan=0, posinf=0, neginf=0)
        
        return text, all_features, cola_mean


def extract_features(df, audio_path, desc):
    scorer = GrammarScorer()
    features, texts, cola_scores, valid_idx = [], [], [], []
    
    for idx, row in tqdm(df.iterrows(), total=len(df), desc=desc):
        filename = row['filename']
        if not filename.endswith('.wav'):
            filename += '.wav'
        filepath = os.path.join(audio_path, filename)
        
        if not os.path.exists(filepath):
            continue
        
        try:
            text, feat, cola = scorer.extract_all(filepath)
            features.append(feat)
            texts.append(text)
            cola_scores.append(cola)
            valid_idx.append(idx)
        except Exception as e:
            print(f"Error {filename}: {e}")
        
        if idx % 30 == 0:
            gc.collect()
            torch.cuda.empty_cache()
    
    return np.array(features), texts, np.array(cola_scores), valid_idx


print("\nExtracting training features...")
X_train, train_texts, train_cola, train_idx = extract_features(train_df, TRAIN_AUDIO_PATH, "Train")
y_train = train_df.loc[train_idx, 'label'].values
print(f"Train shape: {X_train.shape}")

print(f"\nCoLA score correlation with labels: {pearsonr(train_cola, y_train)[0]:.4f}")

print("\nExtracting test features...")
X_test, test_texts, test_cola, test_idx = extract_features(test_df, TEST_AUDIO_PATH, "Test")
print(f"Test shape: {X_test.shape}")

# Scaling
scaler = RobustScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

# Models
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor
from sklearn.linear_model import Ridge, BayesianRidge, HuberRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

print("\nTraining ensemble...")
n_folds = 5
kfold = KFold(n_splits=n_folds, shuffle=True, random_state=SEED)

models = {
    'xgb1': xgb.XGBRegressor(n_estimators=500, max_depth=3, learning_rate=0.05, subsample=0.8,
                             colsample_bytree=0.8, reg_alpha=1, reg_lambda=2, random_state=SEED),
    'xgb2': xgb.XGBRegressor(n_estimators=500, max_depth=4, learning_rate=0.03, subsample=0.7,
                             colsample_bytree=0.7, reg_alpha=2, reg_lambda=3, random_state=SEED+1),
    'lgb1': lgb.LGBMRegressor(n_estimators=500, max_depth=3, learning_rate=0.05, subsample=0.8,
                              colsample_bytree=0.8, reg_alpha=1, reg_lambda=2, random_state=SEED, verbose=-1),
    'lgb2': lgb.LGBMRegressor(n_estimators=500, max_depth=4, learning_rate=0.03, subsample=0.7,
                              colsample_bytree=0.7, reg_alpha=2, reg_lambda=3, random_state=SEED+1, verbose=-1),
    'gbr': GradientBoostingRegressor(n_estimators=300, max_depth=3, learning_rate=0.05,
                                     subsample=0.8, random_state=SEED),
    'rf': RandomForestRegressor(n_estimators=300, max_depth=6, min_samples_leaf=5, random_state=SEED),
    'et': ExtraTreesRegressor(n_estimators=300, max_depth=6, min_samples_leaf=5, random_state=SEED),
    'ridge': Ridge(alpha=10, random_state=SEED),
    'bayesian': BayesianRidge(),
    'huber': HuberRegressor(epsilon=1.35, alpha=0.01),
    'svr': SVR(C=1, epsilon=0.1, kernel='rbf'),
    'knn': KNeighborsRegressor(n_neighbors=10, weights='distance')
}

all_oof = {name: np.zeros(len(X_train_sc)) for name in models}
all_models = {name: [] for name in models}

for fold, (tr_idx, va_idx) in enumerate(kfold.split(X_train_sc)):
    print(f"\nFold {fold+1}/{n_folds}")
    X_tr, X_va = X_train_sc[tr_idx], X_train_sc[va_idx]
    y_tr, y_va = y_train[tr_idx], y_train[va_idx]
    
    for name, model in models.items():
        import copy
        m = copy.deepcopy(model)
        
        if 'xgb' in name:
            m.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
        elif 'lgb' in name:
            m.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], callbacks=[lgb.early_stopping(50, verbose=False)])
        else:
            m.fit(X_tr, y_tr)
        
        all_oof[name][va_idx] = m.predict(X_va)
        all_models[name].append(m)

print("\n" + "="*60)
print("MODEL RESULTS")
print("="*60)
model_rmse = {}
for name in models:
    rmse = np.sqrt(mean_squared_error(y_train, all_oof[name]))
    pearson = pearsonr(y_train, all_oof[name])[0]
    model_rmse[name] = rmse
    print(f"{name}: RMSE={rmse:.4f}, r={pearson:.4f}")

# Ensemble optimization
def optimize_ensemble(weights, preds, y):
    w = np.abs(weights) / np.sum(np.abs(weights))
    blend = sum(wi * preds[name] for wi, name in zip(w, preds.keys()))
    return np.sqrt(mean_squared_error(y, blend))

init_w = np.ones(len(models)) / len(models)
result = minimize(optimize_ensemble, init_w, args=(all_oof, y_train), method='Nelder-Mead', 
                 options={'maxiter': 5000})
opt_w = np.abs(result.x) / np.sum(np.abs(result.x))

print("\nOptimized weights:")
for name, w in zip(models.keys(), opt_w):
    if w > 0.01:
        print(f"  {name}: {w:.4f}")

final_oof = sum(w * all_oof[name] for w, name in zip(opt_w, models.keys()))
final_rmse = np.sqrt(mean_squared_error(y_train, final_oof))
final_pearson = pearsonr(y_train, final_oof)[0]
final_spearman = spearmanr(y_train, final_oof)[0]

print(f"\n" + "="*60)
print("ENSEMBLE RESULTS")
print("="*60)
print(f"OOF RMSE: {final_rmse:.4f}")
print(f"OOF Pearson: {final_pearson:.4f}")
print(f"OOF Spearman: {final_spearman:.4f}")
print("="*60)

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0,0].scatter(y_train, final_oof, alpha=0.6, c=train_cola, cmap='RdYlGn', edgecolors='k', linewidth=0.5)
axes[0,0].plot([1, 5], [1, 5], 'r--', lw=2)
axes[0,0].set_xlabel('Actual Score')
axes[0,0].set_ylabel('Predicted Score')
axes[0,0].set_title(f'OOF (RMSE={final_rmse:.4f}, r={final_pearson:.4f})')
axes[0,0].set_xlim([0.5, 5.5])
axes[0,0].set_ylim([0.5, 5.5])

residuals = y_train - final_oof
axes[0,1].hist(residuals, bins=25, edgecolor='black', alpha=0.7)
axes[0,1].axvline(0, color='r', linestyle='--')
axes[0,1].set_xlabel('Residual')
axes[0,1].set_title(f'Residuals (mean={np.mean(residuals):.3f}, std={np.std(residuals):.3f})')

sorted_models = sorted(model_rmse.items(), key=lambda x: x[1])
names, rmses = zip(*sorted_models)
axes[1,0].barh(names, rmses, color='steelblue', edgecolor='black')
axes[1,0].set_xlabel('RMSE')
axes[1,0].set_title('Model Comparison')
axes[1,0].axvline(final_rmse, color='r', linestyle='--', label=f'Ensemble: {final_rmse:.4f}')
axes[1,0].legend()

axes[1,1].scatter(train_cola, y_train, alpha=0.6, edgecolors='k', linewidth=0.5)
axes[1,1].set_xlabel('CoLA Grammar Score')
axes[1,1].set_ylabel('Actual Label')
axes[1,1].set_title(f'CoLA vs Labels (r={pearsonr(train_cola, y_train)[0]:.4f})')

plt.tight_layout()
plt.savefig('results.png', dpi=150)
plt.show()

# Generate predictions
print("\nGenerating test predictions...")
test_preds = {name: np.mean([m.predict(X_test_sc) for m in all_models[name]], axis=0) for name in models}
final_test = np.clip(sum(w * test_preds[name] for w, name in zip(opt_w, models.keys())), 1, 5)

submission = pd.DataFrame({
    'filename': test_df.loc[test_idx, 'filename'].values,
    'label': final_test
})
submission.to_csv('submission.csv', index=False)
print(f"Saved submission.csv ({len(submission)} samples)")
print(f"Predictions: mean={final_test.mean():.2f}, std={final_test.std():.2f}")

print("\nSample outputs:")
for i in np.random.choice(len(train_texts), 3, replace=False):
    print(f"\nActual: {y_train[i]:.1f} | Pred: {final_oof[i]:.2f} | CoLA: {train_cola[i]:.3f}")
    print(f"Text: {train_texts[i][:100]}...")

print(f"\n{'='*60}")
print(f"FINAL OOF RMSE: {final_rmse:.4f}")
print(f"{'='*60}")
```
